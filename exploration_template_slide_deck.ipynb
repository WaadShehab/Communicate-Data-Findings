```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import os 
import glob
%matplotlib inline


```

```python
df_new=pd.read_csv('201902-fordgobike-tripdata_cleaning.csv')
df_new.shape
```

### When are most trips taken in terms of time of day?



### Thursday is the most day users use bikes in

```python
order_days=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
base_color=sb.color_palette()[0]
plt.figure(figsize=[ 11.69, 8.27])
sb.countplot(data=df_new,x='day',color=base_color,order=order_days);
plt.title('The Distribution Of The Most Day Users Use Bike',fontsize=18);
plt.xticks(rotation=90);
plt.xlabel('Days',fontsize=18);
```

### users use bikes the most from 7 AM to 7 PM.

```python
base_color=sb.color_palette()[0]
plt.figure(figsize=[ 11.69, 8.27])
sb.countplot(data=df_new,x='hour',color=base_color);
plt.title('The Distribution Of The Most Hour Users Use Bike',fontsize=18);
plt.xlabel('Hours',fontsize=18);
```

### Males use bike most

```python
plt.figure(figsize=[ 11.69, 8.27])
sb.countplot(data=df_new, x='member_gender');
plt.title('The Distribution Of Users Age',fontsize=18);
plt.xlabel('Member Gender',fontsize=18);
```

### Subscriber are the most use bikes all over dayes.

```python
order_days=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
plt.figure(figsize=[ 11.69, 8.27])
sb.countplot(data=df_new,x='day',hue='user_type',order=order_days);
plt.xlabel('Dayes Per User Type',fontsize=18);
plt.title('Distribution of User Type Per Dayes',fontsize=18);
plt.xticks(rotation=90);
```

### Female customer go for trip duration longer than female subscriber also male customer go for trip duration longer than male subscriber.

```python
plt.figure(figsize=[ 11.69, 8.27])
sb.barplot(data=df_new,x='user_type',y='duration_sec',hue='member_gender');
plt.title('The Distribution Of Users Types, Users Gender And Duration In Sec',fontsize=18);
plt.xlabel('User Types',fontsize=18);
plt.ylabel('Duration In Secound',fontsize=18);
```

### Conclusion
   - Users use ford bike most on Thursday.
   - Users use ford bike most at 7AM and 7PM.
   - Males who use ford bike more than female.
   - Subscriber are the most use bikes all over dayes.
   - Female customer go for trip duration longer than female subscriber also male customer go for trip duration longer than male subscriber.
